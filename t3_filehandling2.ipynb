```python
#WAP to read the wordlist text file. find the words and there average which have vowels.
f = open("exer/wordlist.txt", 'r')
cnt = 1
vows = set("aeiou")
line = f.readline()
no_of_vows = 0
while line:
    lset = set(line)
    if lset.issuperset(vows):
        no_of_vows +=1
    cnt+=1
    line = f.readline()
else:
    print(line)

print(cnt)
print(no_of_vows)
print("avg:", no_of_vows/cnt)
f.close()
```

```python
# WAP to find the longest word that contains abcdef in anyorder
f = open("exer/wordlist.txt", 'r')
cnt = 1
wrds = set("abcdef")
line = f.readline()
longest = ""
while line:
    lset = set(line)
    if lset.issuperset(wrds):
        if len(longest)<len(line):
            longest = line
    line = f.readline()
else:
    print(longest)

f.close()
```
